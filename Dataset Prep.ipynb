```python
# Dataset containing the learning system data 
ratingdata = pd.read_csv('C:/Users/Avani/Documents/MScProj/Data/user_course_score.csv')
ratingdata.head()
```

```python
#making 2 copies of the data for preprocessing
c1_ratingdata = ratingdata.copy()
c2_ratingdata = ratingdata.copy()
```

```python
# Some operations to make the dataset more comprehensive
c1 = c1_ratingdata.drop(['skill_mod2','skill_mod2_level','skill1_rank','skill2_rank',
                         'course_name2','course_name2_rating','course_name2_doc','course_name2_id'], axis=1)

c2 = c2_ratingdata.drop(['skill_mod1','skill_mod1_level','skill1_rank','skill2_rank','course_name1',
                         'course_name1_rating','course_name1_doc','course_name1_id'], axis=1)


c1 = c1.rename(columns={'skill_mod1':'skill_mod','skill_mod1_level':'skill_mod_level','course_name1':'course_name','course_name1_rating':'course_rating','course_name1_doc':'course_doc','course_name1_id':'course_id'})
c2 = c2.rename(columns={'skill_mod2':'skill_mod','skill_mod2_level':'skill_mod_level','course_name2':'course_name','course_name2_rating':'course_rating','course_name2_doc':'course_doc','course_name2_id':'course_id'})
```

```python
c1.head()
```

```python
c2.head()
```

```python
#combining the data again to make a single dataframe
ratingdata = pd.concat([c1,c2])
```

```python
# dropping some more redundant attributes
ratingdata = ratingdata.drop(['course_doc','skill_mod','skill_mod_level','time_per_week'],axis=1)
```

```python
# Checking for duplicates 
duplicate_rating_data = ratingdata.duplicated().sum()
duplicate_rating_data
```

```python
ratingdata = ratingdata.drop_duplicates()
shape_rating_data = ratingdata.shape
shape_rating_data
```
