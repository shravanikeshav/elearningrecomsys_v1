```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
rating_data = pd.read_csv('C:/Users/Avani/Documents/MScProj/Data/ratingdata.csv')
rating_data.head()
```

#### Distribution of Ratings: 
Understand the overall distribution of ratings given by employees.

```python
# Distribution of Ratings
plt.figure(figsize=(6, 4))
sns.countplot(x='course_rating', data=rating_data, palette='pastel')
plt.title('Distribution of Ratings')
plt.xlabel('Course Rating')
plt.ylabel('Frequency')
plt.tight_layout()
plt.show()
```

Distribution of Ratings:

Ratings of 5 are the most frequent, indicating that many users are highly satisfied with the courses they took.
Ratings of 4 are also common, suggesting general satisfaction.
Ratings of 1, 2, and 3 are less frequent compared to 4 and 5, with ratings of 3 being the least common.

```python
# Ratings by Employee Status
plt.figure(figsize=(6, 4))
sns.countplot(x='course_rating', hue='emp_status', data=rating_data, palette='muted')
plt.title('Distribution of Ratings by Employee Status')
plt.xlabel('Course Rating')
plt.ylabel('Frequency')
plt.legend(title='Employee Status')
plt.tight_layout()
plt.show()
```

Distribution of Ratings by Employee Status:

Employees with the "Active" status predominantly give ratings of 5, followed by ratings of 4.
Employees in the "Transition In" status show a similar pattern, with ratings of 5 being the most common.
The "Transition Out" group has fewer ratings overall, but the pattern remains consistent with the other groups, showing a preference for ratings of 5.

```python
# Top rated courses based on 5-star ratings
top_rated_courses = rating_data[rating_data['course_rating'] == 5]['course_name'].value_counts().head(10)

plt.figure(figsize=(6,4))
top_rated_courses.plot(kind='barh', color='salmon')
plt.title('Top 10 Courses by 5-Star Ratings')
plt.xlabel('Number of 5-Star Ratings')
plt.ylabel('Course Name')
plt.tight_layout()
plt.show()
```

Top 10 Courses by 5-Star Ratings:

The chart displays the courses that have received the highest number of 5-star ratings.
Some courses stand out with a significantly higher number of top ratings, indicating their popularity and user satisfaction.

```python
# Average rating for each employee path
path_avg_rating = rating_data.groupby('emp_path')['course_rating'].mean().sort_values(ascending=False)

plt.figure(figsize=(6,4))
path_avg_rating.plot(kind='bar', color='teal')
plt.title('Average Rating by Employee Path')
plt.xlabel('Employee Path')
plt.ylabel('Average Rating')
plt.tight_layout()
plt.show()
```

Average Rating by Employee Path:

Most employee paths have average ratings that hover around the 4 to 5 range.
Some paths, like DS and Big Data, have slightly higher average ratings, indicating that employees in these paths might be more satisfied with the courses they've taken.
Conversely, paths like Cloud have a slightly lower average, although the difference is not stark.
