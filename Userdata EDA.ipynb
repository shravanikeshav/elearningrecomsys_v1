```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
user_data = pd.read_csv('C:/Users/Avani/Documents/MScProj/Data/userdata.csv')
user_data.head()
```

```python
missing_user_data = user_data.isnull().sum()
missing_user_data
```

```python
fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(10,4))
# Distribution of edu_level in user data
sns.countplot(y= user_data['edu_level'], ax=axes[0])
axes[0].set_title('Distribution of Education Level (User Data)')
axes[0].set_xlabel('Frequency')
axes[0].set_ylabel('Education Level')

# Distribution of emp_status in user data
sns.countplot(y= user_data['emp_status'], ax=axes[1])
axes[1].set_title('Distribution of Employee Status (User Data)')
axes[1].set_xlabel('Frequency')
axes[1].set_ylabel('Employee Status')

plt.tight_layout()
plt.show()
```

##### Distribution of Education Level (User Data):
Most of the employees have a Bachelor's degree, followed by those with a Master's. The number of employees with a PhD is relatively lower.
##### Distribution of Employee Status (User Data):
A significant number of employees are in the "Transition In" status, followed closely by those who are "Active". There are fewer employees in the "Transition Out" status.

```python
# Correlation matrix for user data
correlation_matrix_user = user_data[['emp_level', 'time_per_week']].corr()

# Visualizing the correlation matrix using a heatmap
plt.figure(figsize=(6, 4))
sns.heatmap(correlation_matrix_user, annot=True, cmap='coolwarm', vmin=-1, vmax=1)
plt.title('Correlation Heatmap (User Data)')
plt.show()
```

emp_level and time_per_week: There's a very weak correlation between the employee level and the time they spend per week on courses. This suggests that as employees progress in levels, there might or might not be a change in the time they allocate for courses each week. However, the correlation is very weak, so it's not a strong relationship.

```python
fig, axes = plt.subplots(nrows=1, ncols=3, figsize=(18, 6))

# Distribution of time_per_week across edu_level
sns.boxplot(x='edu_level', y='time_per_week', data=user_data, ax=axes[0])
axes[0].set_title('Distribution of Time per Week by Education Level')
axes[0].set_xlabel('Education Level')
axes[0].set_ylabel('Time per Week')

# Distribution of time_per_week across emp_status
sns.boxplot(x='emp_status', y='time_per_week', data=user_data, ax=axes[1])
axes[1].set_title('Distribution of Time per Week by Employee Status')
axes[1].set_xlabel('Employee Status')
axes[1].set_ylabel('Time per Week')

# Boxplot to visualize time_per_week across emp_level
sns.boxplot(x='emp_level', y='time_per_week', data=user_data, ax=axes[2])
axes[2].set_title('Distribution of Time per Week by Employee Level')
axes[2].set_xlabel('Employee Level')
axes[2].set_ylabel('Time per Week')

plt.tight_layout()
plt.show()
```

#### Distribution of Time per Week by Education Level:

Employees with a PhD tend to spend slightly more time per week on courses compared to other education levels, as indicated by the higher median value.
Employees with a Bachelor's degree show a wider spread in the time they spend per week, while those with a Master's degree have a more concentrated distribution.

#### Distribution of Time per Week by Employee Status:

Employees in the "Active" and "Transition In" statuses have similar distributions of time spent per week on courses.
Employees in the "Transition Out" status seem to spend slightly less time on courses, as indicated by the lower median value.

#### Distribution of Time per Week by Employee Level:

There's a trend where as the employee level increases, the median time spent on courses per week slightly decreases. However, higher employee levels (e.g., 6 and 7) have wider spreads, indicating more variability in the time spent among these employees.
